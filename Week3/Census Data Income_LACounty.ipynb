```python
print('renaming census tracts to match income range')
```

```python
gdf.columns = ['geoid',
              'name',
              'total',
              'Less than $10,000',
              '$10,000 to $14,999',
              '$15,000 to $19,999',
              '$20,000 to $24,999',
              '$25,000 to $29,999',
              '$30,000 to $34,999',
              '$35,000 to $39,999',
              '$40,000 to $44,999',
              '$45,000 to $49,999',
              '$50,000 to $59,999',
              '$60,000 to $74,999',
              'geometry']
```

```python
gdf.head()
```

```python
gdf = gdf.drop([0])
```

```python
print('creating percent categories')
```

```python
gdf['Percent Less than $10,000'] + gdf['Percent $10,000 to $14,999'] + gdf['Percent $15,000 to $19,999'] + gdf['Percent $20,000 to $24,999']
```

```python
gdf.sample(5)
```

```python
print('adding a new category combining the four lowest income categories together. I modelled the category after poverty levels of income of a family of three.')
```

```python
gdf['Percent Poverty'] = gdf['Percent Less than $10,000'] + gdf['Percent $10,000 to $14,999'] + gdf['Percent $15,000 to $19,999'] + gdf['Percent $20,000 to $24,999']
```

```python
gdf.sample(5)
```

```python
print('map of census tracts percentage of poverty')
```

```python
gdf.plot(figsize=(12,10),
                 column='Percent Poverty',
                 legend=True, 
                 cmap='RdPu',
                 scheme='NaturalBreaks')
```

```python
gdf[gdf['Percent Poverty'] > 25]
```

```python
print('map of census tracts where at least 1/3 of people are making $24,999 or less')
```

```python
gdf[gdf['Percent Poverty'] > 33.3].plot(figsize=(12,10))
```

```python
gdf_web_mercator = gdf.to_crs(epsg=3857)
```

```python
gdf_web_mercator.head()
```

```python
print('map of census tracts where at least 1/3 of people are making $24,999 or less with basemap')
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
gdf_web_mercator[gdf_web_mercator['Percent Poverty'] > 33.3].plot(ax=ax, alpha=0.8)
ax.axis('off')
ax.set_title('Census Tracts with more than 33% of People Living in Poverty (Less than $24,999 for a Household of Three)',fontsize=24)
ctx.add_basemap(ax)
```
